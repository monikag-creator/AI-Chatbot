```python
#  Study Assistant Chatbot

def get_response(user_input):
    #Return a response based on keywords in user input.
    user_input = user_input.lower().strip()

    # Greeting
    if any(word in user_input for word in ["hello", "hi", "hey", "greetings"]):
        return (
            "Hello!  I'm your Study Assistant Bot.\n"
            "  I'm here to help you with study tips, exam prep, time management, and more!\n"
            "  What would you like to know?"
        )

    # Study Tips
    elif any(word in user_input for word in ["study tips", "study", "how to study", "study advice"]):
        return (
            " * Here are some effective Study Tips:\n"
            "  1. Break your study sessions into 25-min chunks (Pomodoro Technique).\n"
            "  2. Use active recall — test yourself instead of just re-reading.\n"
            "  3. Summarize topics in your own words.\n"
            "  4. Study in a quiet, distraction-free environment.\n"
            "  5. Review your notes within 24 hours of learning."
        )

    # Exam Preparation
    elif any(word in user_input for word in ["exam", "exam preparation", "test", "prepare", "revision"]):
        return (
            " * Exam Preparation Tips:\n"
            "  1. Start revising at least 1 week before the exam.\n"
            "  2. Practice past papers and sample questions.\n"
            "  3. Focus on weak areas first.\n"
            "  4. Make short notes or flashcards for quick revision.\n"
            "  5. Get a good night's sleep before the exam — rest is crucial!"
        )

    # Time Management
    elif any(word in user_input for word in ["time management", "time", "schedule", "plan", "organize"]):
        return (
            " * Time Management Tips:\n"
            "  1. Create a daily study timetable and stick to it.\n"
            "  2. Prioritize tasks — do important ones first.\n"
            "  3. Avoid multitasking; focus on one subject at a time.\n"
            "  4. Set specific goals for each study session.\n"
            "  5. Take short breaks — it boosts productivity!"
        )

    # Motivation
    elif any(word in user_input for word in ["motivation", "motivate", "inspired", "discouraged", "give up", "tired"]):
        return (
            " * Stay Motivated:\n"
            "  1. Remember your WHY — why did you start?\n"
            "  2. Celebrate small wins — every page learned counts!\n"
            "  3. Surround yourself with positive, hardworking people.\n"
            "  4. Visualize your success — imagine achieving your goal.\n"
            "  5. Quote: 'Success is the sum of small efforts repeated daily.' "
        )

    # Help / Options
    elif any(word in user_input for word in ["help", "options", "what can you do", "topics"]):
        return (
            "  * I can help you with:\n"
            "  • study tips\n"
            "  • exam preparation\n"
            "  • time management\n"
            "  • motivation\n"
            "  Just type any of these topics!"
        )

    # Default response
    else:
        return (
            " * Sorry, I didn't understand that.\n"
            "  Try asking about: study tips, exam preparation, time management, or motivation.\n"
            "  Type 'help' to see all options."
        )

 # Greeting banner
def main():
    print("=" * 55)
    print("  WELCOME TO STUDY ASSISTANT CHATBOT ")
    print("=" * 55)
    print("  I'm here to help you study smarter and stay motivated!")
    print("  Type 'bye!' or 'exit' to end the conversation.")
    print("=" * 55)
    print()

    # Main chat loop
    while True:
        user_input = input("You: ").strip()

        # Skip empty input
        if not user_input:
            print("Bot: Please type something! Type 'help' to see options.\n")
            continue

        # Exit condition
        if user_input.lower() in ["bye!", "exit!", "quit", "goodbye!"]:
            print("Bot: Goodbye! Best of luck with your studies. You've got this! ")
            print("=" * 55)
            break

        # Get and print response
        response = get_response(user_input)
        print(f"Bot: {response}\n")


# Run the chatbot
if __name__ == "__main__":
    main()
```
